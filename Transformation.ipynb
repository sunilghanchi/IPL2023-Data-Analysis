```python
import pandas as pd
```

```python
df1 = pd.read_csv("D:\Data Analysis\Projects\IPL2023\data\IPLdata.csv")
```

```python
teams = {
    "Titans":'Gujarat Titans',
    "Super Kings":'Chennai Super Kings',
    "Punjab Kings":'Punjab Kings',
    "KKR":'Kolkata Knight Riders',
    "Super Giants":'Lucknow Super Giants',
    "Capitals":'Delhi Capitals',
    "Sunrisers":'Sunrisers Hyderabad',
    "Royals":'Rajasthan Royals',
    "RCB":'Royal Challengers Bangalore',
    "Mumbai":'Mumbai Indians',
}
```

```python
df1['Team 1'] = df1['Team 1'].map(teams)
df1['Team 2'] = df1['Team 2'].map(teams)
```

```python
df1['Winner'] = df1['Winner'].map(teams)
```

```python
df1.drop('Team', axis=1)
```

```python
df1['Team'] = df1['Team 1']+' Vs '+ df1['Team 2']
```

```python
df1
```

```python
df1 = df1.drop('Scoreboard Link', axis=1)
```

```python
df1.isnull().sum()
```

```python
df1
```

```python
df1.to_csv("D:\Data Analysis\Projects\IPL2023\IPLSchedule.csv", index=False)
```
